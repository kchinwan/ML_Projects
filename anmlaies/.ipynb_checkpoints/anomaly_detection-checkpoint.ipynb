```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.ensemble import IsolationForest
import warnings
from prophet import Prophet
```

```python
warnings.filterwarnings('ignore')
```

```python
df_one = pd.read_csv("processed_write_ops_db.csv")
```

```python
df_one = df_one.dropna()
```

```python
df_model = df_one.copy()
```

```python
# Train IsolationForest
model = IsolationForest(contamination=0.01, random_state=42)  
df_one["anomaly_score"] = model.fit_predict(df_one[["write_ops_norm"]])
```

```python
# Map predictions: -1 = anomaly, 1 = normal
df_one["is_anomaly"] = df_one["anomaly_score"].apply(lambda x: 1 if x == -1 else 0)
```

```python
# Plot results
plt.figure(figsize=(15,6))
plt.plot(df_one["timestamp"], df_one["write_ops_norm"], label="write_ops_norm", color="blue")
plt.scatter(df_one.loc[df_one["is_anomaly"]==1, "timestamp"],
            df_one.loc[df_one["is_anomaly"]==1, "write_ops_norm"],
            color="red", marker="x", label="Anomaly")
plt.title(f"IsolationForest Anomaly Detection for ")
plt.xlabel("Time")
plt.ylabel("Normalized Write Ops")
plt.legend()
plt.show()
```

```python
# Prophet expects specific column names
df_model["timestamp"] = pd.to_datetime(df_model["timestamp"])
df_model.rename(columns={"timestamp": "ds", "write_ops": "y"}, inplace=True)
```

```python
# Initialize model with daily seasonality
model = Prophet(
    daily_seasonality=True,
    weekly_seasonality=False,   # You only have 1 week for now
    yearly_seasonality=False
)

model.fit(df_model)
```

```python
# Forecast same horizon (for detection) or future (for prediction)
future = model.make_future_dataframe(periods=60, freq="min")  # predict next 60 mins
forecast = model.predict(future)
```

```python
# Merge forecast with actual values
df_out = df_model.merge(
    forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]],
    on="ds", how="left"
)

# Residuals = actual - forecast
df_out["residual"] = df_out["y"] - df_out["yhat"]

# Simple anomaly rule: outside confidence interval
df_out["anomaly"] = (df_out["y"] > df_out["yhat_upper"]) | (df_out["y"] < df_out["yhat_lower"])
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_out["ds"], df_out["y"], label="Actual", alpha=0.6)
plt.plot(df_out["ds"], df_out["yhat"], label="Forecast", color="orange")
plt.fill_between(df_out["ds"], df_out["yhat_lower"], df_out["yhat_upper"], color="orange", alpha=0.2)

# Highlight anomalies
anoms = df_out[df_out["anomaly"]]
plt.scatter(anoms["ds"], anoms["y"], color="red", label="Anomaly", marker="x")

plt.legend()
plt.show()
```
